# **Vehicle Detection System**
Uses a YOLO model to detect vehicles in frames.

```python
# PHASE 1: VEHICLE DETECTION SYSTEM

class VehicleDetector:
    """Vehicle detection using YOLO model"""

    def __init__(self, model_name='yolov8n.pt', model_path=None):
        """Initialize the detector with YOLO model"""
        print("Loading YOLO model...")
        if model_path:
            self.model = YOLO(model_path)
            print(f"Loaded model from: {model_path}")
        else:
            self.model = YOLO(model_name)
            print(f"Loaded default model: {model_name}")

        self.vehicle_classes = [2, 3, 5, 7]  # car, motorcycle, bus, truck in COCO dataset
        print("Vehicle detector initialized!")

    def detect_vehicles(self, frame):
        """Detect vehicles in a single frame"""
        results = self.model(frame, verbose=False)
        detections = []

        for result in results:
            boxes = result.boxes
            if boxes is not None:
                for i, box in enumerate(boxes):
                    class_id = int(box.cls[0])
                    if class_id in self.vehicle_classes:
                        x1, y1, x2, y2 = box.xyxy[0].cpu().numpy()
                        confidence = float(box.conf[0])

                        detections.append({
                            'bbox': [int(x1), int(y1), int(x2), int(y2)],
                            'confidence': confidence,
                            'class_id': class_id,
                            'center': [(x1 + x2) / 2, (y1 + y2) / 2]
                        })

        return detections

    def draw_detections(self, frame, detections):
        """Draw bounding boxes on frame"""
        annotated_frame = frame.copy()

        for detection in detections:
            x1, y1, x2, y2 = detection['bbox']
            confidence = detection['confidence']

            # Draw bounding box
            cv2.rectangle(annotated_frame, (x1, y1), (x2, y2), (0, 255, 0), 2)

            # Add label
            label = f"Vehicle: {confidence:.2f}"
            cv2.putText(annotated_frame, label, (x1, y1 - 10),
                       cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

        return annotated_frame
```

# **Vehicle Tracking System**
Tracks vehicles across frames using centroid-based tracking.

```python
# PHASE 2: VEHICLE TRACKING SYSTEM

class VehicleTracker:
    """Track vehicles across frames"""

    def __init__(self, max_disappeared=30, max_distance=100):
        self.next_object_id = 0
        self.objects = {}
        self.disappeared = {}
        self.max_disappeared = max_disappeared
        self.max_distance = max_distance

    def register(self, centroid):
        """Register a new object"""
        self.objects[self.next_object_id] = centroid
        self.disappeared[self.next_object_id] = 0
        self.next_object_id += 1

    def deregister(self, object_id):
        """Remove an object from tracking"""
        del self.objects[object_id]
        del self.disappeared[object_id]

    def update(self, detections):
        """Update tracker with new detections"""
        if len(detections) == 0:
            # Mark existing objects as disappeared
            for object_id in list(self.disappeared.keys()):
                self.disappeared[object_id] += 1
                if self.disappeared[object_id] > self.max_disappeared:
                    self.deregister(object_id)
            return {}

        input_centroids = np.array([det['center'] for det in detections])

        if len(self.objects) == 0:
            for centroid in input_centroids:
                self.register(centroid)
        else:
            object_centroids = list(self.objects.values())
            object_ids = list(self.objects.keys())

            # Compute distance matrix
            D = np.linalg.norm(np.array(object_centroids)[:, np.newaxis] - input_centroids, axis=2)

            # Find minimum values and sort by distance
            rows = D.min(axis=1).argsort()
            cols = D.argmin(axis=1)[rows]

            used_row_indices = set()
            used_col_indices = set()

            for (row, col) in zip(rows, cols):
                if row in used_row_indices or col in used_col_indices:
                    continue

                if D[row, col] <= self.max_distance:
                    object_id = object_ids[row]
                    self.objects[object_id] = input_centroids[col]
                    self.disappeared[object_id] = 0

                    used_row_indices.add(row)
                    used_col_indices.add(col)

            unused_row_indices = set(range(0, D.shape[0])).difference(used_row_indices)
            unused_col_indices = set(range(0, D.shape[1])).difference(used_col_indices)

            if D.shape[0] >= D.shape[1]:
                for row in unused_row_indices:
                    object_id = object_ids[row]
                    self.disappeared[object_id] += 1

                    if self.disappeared[object_id] > self.max_disappeared:
                        self.deregister(object_id)
            else:
                for col in unused_col_indices:
                    self.register(input_centroids[col])

        return self.objects.copy()
```

# **Traffic Flow Analyzer**
Analyzes traffic patterns, calculates flow rate, and detects congestion.

```python
# PHASE 2: TRAFFIC FLOW ANALYZER

class TrafficFlowAnalyzer:
    """Analyze traffic flow patterns"""

    def __init__(self):
        self.vehicle_counts = deque(maxlen=60)  # Store last 60 seconds
        self.speed_data = []
        self.congestion_threshold = 0.7
        self.flow_history = defaultdict(list)

    def update_vehicle_count(self, count):
        """Update vehicle count with timestamp"""
        timestamp = datetime.now()
        self.vehicle_counts.append({
            'timestamp': timestamp,
            'count': count
        })

    def calculate_flow_rate(self):
        """Calculate vehicles per minute"""
        if len(self.vehicle_counts) < 2:
            return 0

        recent_counts = list(self.vehicle_counts)[-10:]  # Last 10 seconds
        if len(recent_counts) < 2:
            return 0

        time_diff = (recent_counts[-1]['timestamp'] - recent_counts[0]['timestamp']).total_seconds()
        if time_diff == 0:
            return 0

        count_diff = recent_counts[-1]['count'] - recent_counts[0]['count']
        flow_rate = (count_diff / time_diff) * 60  # vehicles per minute

        return max(0, flow_rate)

    def detect_congestion(self, current_density, average_density):
        """Detect if there's congestion"""
        if average_density == 0:
            return False

        congestion_ratio = current_density / average_density
        return congestion_ratio > self.congestion_threshold

    def get_traffic_metrics(self):
        """Get comprehensive traffic metrics"""
        flow_rate = self.calculate_flow_rate()
        current_count = self.vehicle_counts[-1]['count'] if self.vehicle_counts else 0

        # Calculate average density over last 5 minutes
        five_min_ago = datetime.now() - timedelta(minutes=5)
        recent_data = [entry for entry in self.vehicle_counts
                      if entry['timestamp'] >= five_min_ago]

        avg_density = np.mean([entry['count'] for entry in recent_data]) if recent_data else 0

        is_congested = self.detect_congestion(current_count, avg_density)

        return {
            'current_vehicles': current_count,
            'flow_rate': flow_rate,
            'average_density': avg_density,
            'is_congested': is_congested,
            'congestion_level': min(current_count / max(avg_density, 1), 2.0)
        }
```

# **Traffic Signal Optimizer**
Optimizes traffic signal timings based on real-time traffic data.

```python
# PHASE 3: SIGNAL TIMING OPTIMIZER

class TrafficSignalOptimizer:
    """Optimize traffic signal timings"""

    def __init__(self):
        self.min_green_time = 15  # Minimum green light duration
        self.max_green_time = 120  # Maximum green light duration
        self.yellow_time = 3
        self.red_clearance = 2

        # Default timing plan
        self.default_timings = {
            'north_south_green': 45,
            'east_west_green': 45,
            'cycle_length': 100
        }

    def calculate_optimal_timing(self, traffic_data):
        """Calculate optimal signal timing based on traffic data"""
        ns_demand = traffic_data.get('north_south_flow', 0)
        ew_demand = traffic_data.get('east_west_flow', 0)

        total_demand = ns_demand + ew_demand

        if total_demand == 0:
            return self.default_timings

        # Proportional allocation based on demand
        ns_ratio = ns_demand / total_demand
        ew_ratio = ew_demand / total_demand

        # Base cycle length adjustment for congestion
        base_cycle = 90
        congestion_factor = traffic_data.get('congestion_level', 1.0)
        adjusted_cycle = min(base_cycle * congestion_factor, 150)

        # Calculate green times (reserve time for yellow and clearance)
        available_green_time = adjusted_cycle - 2 * (self.yellow_time + self.red_clearance)

        ns_green = max(self.min_green_time,
                      min(self.max_green_time, available_green_time * ns_ratio))
        ew_green = max(self.min_green_time,
                      min(self.max_green_time, available_green_time * ew_ratio))

        # Ensure total doesn't exceed available time
        total_green = ns_green + ew_green
        if total_green > available_green_time:
            scale_factor = available_green_time / total_green
            ns_green *= scale_factor
            ew_green *= scale_factor

        return {
            'north_south_green': int(ns_green),
            'east_west_green': int(ew_green),
            'cycle_length': int(ns_green + ew_green + 2 * (self.yellow_time + self.red_clearance)),
            'yellow_time': self.yellow_time,
            'red_clearance': self.red_clearance
        }

    def simulate_performance(self, timings, traffic_data):
        """Simulate performance of given timing plan"""
        # Simple performance metric based on delay estimation
        cycle_length = timings['cycle_length']
        ns_green_ratio = timings['north_south_green'] / cycle_length
        ew_green_ratio = timings['east_west_green'] / cycle_length

        ns_flow = traffic_data.get('north_south_flow', 0)
        ew_flow = traffic_data.get('east_west_flow', 0)

        # Calculate capacity utilization
        ns_capacity_ratio = (ns_flow / 60) / ns_green_ratio if ns_green_ratio > 0 else 0
        ew_capacity_ratio = (ew_flow / 60) / ew_green_ratio if ew_green_ratio > 0 else 0

        # Simple delay model (Webster's formula approximation)
        ns_delay = (cycle_length * (1 - ns_green_ratio)**2) / (2 * (1 - ns_capacity_ratio)) if ns_capacity_ratio < 1 else float('inf')
        ew_delay = (cycle_length * (1 - ew_green_ratio)**2) / (2 * (1 - ew_capacity_ratio)) if ew_capacity_ratio < 1 else float('inf')

        avg_delay = (ns_delay + ew_delay) / 2

        return {
            'average_delay': avg_delay,
            'ns_capacity_ratio': ns_capacity_ratio,
            'ew_capacity_ratio': ew_capacity_ratio,
            'performance_score': 100 - min(avg_delay, 100)  # Higher is better
        }
```

# **Complete Traffic Management System**
This class integrates the detection, tracking, analysis, and optimization components to provide a complete traffic management solution.

```python
# PHASE 4: COMPLETE TRAFFIC MANAGEMENT SYSTEM

class TrafficManagementSystem:
    """Complete traffic management system"""

    def __init__(self, model_path=None):
        # Initialize the VehicleDetector. Use a pre-trained model for demonstration.
        # trained_model_path = 'runs/detect/train3/weights/best.pt' # Path to the model trained on dummy data
        self.detector = VehicleDetector(model_name='yolov8n.pt') # Use a pre-trained model
        self.tracker = VehicleTracker()
        self.analyzer = TrafficFlowAnalyzer()
        self.optimizer = TrafficSignalOptimizer()

        # System state
        self.is_running = False
        self.current_frame = None
        self.system_metrics = {}

        # Data storage
        self.setup_database()

    def setup_database(self):
        """Setup SQLite database for data storage"""
        self.conn = sqlite3.connect(':memory:', check_same_thread=False)
        cursor = self.conn.cursor()

        cursor.execute('''
            CREATE TABLE traffic_data (
                id INTEGER PRIMARY KEY AUTOINCREMENT,
                timestamp DATETIME DEFAULT CURRENT_TIMESTAMP,
                vehicle_count INTEGER,
                flow_rate REAL,
                congestion_level REAL,
                signal_timing TEXT
            )
        ''')

        self.conn.commit()
        print("Database initialized!")

    def process_frame(self, frame):
        """Process a single frame through the complete pipeline"""
        # Detect vehicles
        detections = self.detector.detect_vehicles(frame)

        # Track vehicles
        tracked_objects = self.tracker.update(detections)

        # Update traffic analysis
        vehicle_count = len(tracked_objects)
        self.analyzer.update_vehicle_count(vehicle_count)

        # Get traffic metrics
        traffic_metrics = self.analyzer.get_traffic_metrics()

        # Optimize signal timing
        traffic_data = {
            'north_south_flow': traffic_metrics['flow_rate'] * 0.6,  # Simulated directional split
            'east_west_flow': traffic_metrics['flow_rate'] * 0.4,
            'congestion_level': traffic_metrics['congestion_level']
        }

        optimal_timings = self.optimizer.calculate_optimal_timing(traffic_data)
        performance = self.optimizer.simulate_performance(optimal_timings, traffic_data)

        # Store data in database
        self.store_data(traffic_metrics, optimal_timings)

        # Draw annotations
        annotated_frame = self.detector.draw_detections(frame, detections)

        # Add system info overlay
        self.add_info_overlay(annotated_frame, traffic_metrics, optimal_timings, performance)

        return annotated_frame, traffic_metrics, optimal_timings

    def add_info_overlay(self, frame, traffic_metrics, timings, performance):
        """Add information overlay to frame"""
        h, w = frame.shape[:2]

        # Create semi-transparent overlay
        overlay = frame.copy()

        # Traffic info
        info_text = [
            f"Vehicles: {traffic_metrics['current_vehicles']}",
            f"Flow: {traffic_metrics['flow_rate']:.1f} veh/min",
            f"Congestion: {'YES' if traffic_metrics['is_congested'] else 'NO'}",
            f"Level: {traffic_metrics['congestion_level']:.2f}",
            "",
            f"Signal Timings:",
            f"NS Green: {timings['north_south_green']}s",
            f"EW Green: {timings['east_west_green']}s",
            f"Cycle: {timings['cycle_length']}s",
            f"Performance: {performance['performance_score']:.1f}"
        ]

        y_offset = 30
        for text in info_text:
            cv2.putText(overlay, text, (10, y_offset),
                       cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 255, 255), 2)
            y_offset += 25

        # Blend overlay with original frame
        cv2.addWeighted(overlay, 0.8, frame, 0.2, 0, frame)

    def store_data(self, metrics, timings):
        """Store data in database"""
        cursor = self.conn.cursor()
        cursor.execute('''
            INSERT INTO traffic_data
            (vehicle_count, flow_rate, congestion_level, signal_timing)
            VALUES (?, ?, ?, ?)
        ''', (
            metrics['current_vehicles'],
            metrics['flow_rate'],
            metrics['congestion_level'],
            json.dumps(timings)
        ))
        self.conn.commit()
```

```python
def process_uploaded_image():
    """Process an uploaded traffic image"""

    # Create file upload widget
    from google.colab import files
    import cv2
    import numpy as np
    from PIL import Image
    import matplotlib.pyplot as plt

    print("Upload a traffic image (JPG, PNG):")
    uploaded = files.upload()

    for filename, data in uploaded.items():
        # Load image
        image = Image.open(io.BytesIO(data))
        frame = cv2.cvtColor(np.array(image), cv2.COLOR_RGB2BGR)

        # Initialize system and process
        system = TrafficManagementSystem()
        processed_frame, metrics, timings = system.process_frame(frame)

        # Display results
        fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

        # Original image
        ax1.imshow(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
        ax1.set_title("Original Image")
        ax1.axis('off')

        # Processed image with detections
        ax2.imshow(cv2.cvtColor(processed_frame, cv2.COLOR_BGR2RGB))
        ax2.set_title("Vehicle Detection & Analysis")
        ax2.axis('off')

        plt.tight_layout()
        plt.show()

        # Print metrics
        print("\n📊 TRAFFIC ANALYSIS RESULTS:")
        print(f"🚗 Vehicles Detected: {metrics['current_vehicles']}")
        print(f"🚦 Flow Rate: {metrics['flow_rate']:.1f} vehicles/minute")
        print(f"🔴 Congestion Status: {'CONGESTED' if metrics['is_congested'] else 'NORMAL'}")
        print(f"📈 Congestion Level: {metrics['congestion_level']:.2f}")

        print("\n⚡ OPTIMIZED SIGNAL TIMINGS:")
        print(f"🟢 North-South Green: {timings['north_south_green']} seconds")
        print(f"🟢 East-West Green: {timings['east_west_green']} seconds")
        print(f"🔄 Total Cycle Length: {timings['cycle_length']} seconds")

        break

# Run the example
# Uncomment the line below to test with an image upload
# process_uploaded_image()

# =============================================================================
# EXAMPLE 2: SIMULATE REAL-TIME PROCESSING WITH SAMPLE DATA
# =============================================================================

def simulate_realtime_processing():
    """Simulate real-time traffic processing with synthetic data"""

    import time
    import random
    from IPython.display import clear_output
    import matplotlib.pyplot as plt

    print("🚀 Starting Real-Time Traffic Simulation...")
    print("This will simulate 60 seconds of traffic processing\n")

    # Initialize system
    system = TrafficManagementSystem()

    # Storage for plotting
    timestamps = []
    vehicle_counts = []
    flow_rates = []
    congestion_levels = []
    ns_green_times = []
    ew_green_times = []

    start_time = time.time()

    for i in range(60):  # 60 seconds simulation
        current_time = time.time() - start_time

        # Simulate varying traffic conditions
        base_vehicles = 15 + 10 * np.sin(current_time * 0.1) + random.randint(-5, 5)
        base_vehicles = max(0, int(base_vehicles))

        # Create mock traffic data
        system.analyzer.update_vehicle_count(base_vehicles)
        metrics = system.analyzer.get_traffic_metrics()

        # Get optimal timings
        traffic_data = {
            'north_south_flow': metrics['flow_rate'] * (0.4 + 0.2 * np.sin(current_time * 0.05)),
            'east_west_flow': metrics['flow_rate'] * (0.6 - 0.2 * np.sin(current_time * 0.05)),
            'congestion_level': metrics['congestion_level']
        }
        timings = system.optimizer.calculate_optimal_timing(traffic_data)

        # Store data
        timestamps.append(current_time)
        vehicle_counts.append(metrics['current_vehicles'])
        flow_rates.append(metrics['flow_rate'])
        congestion_levels.append(metrics['congestion_level'])
        ns_green_times.append(timings['north_south_green'])
        ew_green_times.append(timings['east_west_green'])

        # Update display every 5 seconds
        if i % 5 == 0:
            clear_output(wait=True)

            fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 10))

            # Vehicle count over time
            ax1.plot(timestamps, vehicle_counts, 'b-', linewidth=2)
            ax1.set_title('Vehicle Count Over Time')
            ax1.set_xlabel('Time (seconds)')
            ax1.set_ylabel('Number of Vehicles')
            ax1.grid(True)

            # Flow rate
            ax2.plot(timestamps, flow_rates, 'g-', linewidth=2)
            ax2.set_title('Traffic Flow Rate')
            ax2.set_xlabel('Time (seconds)')
            ax2.set_ylabel('Vehicles per Minute')
            ax2.grid(True)

            # Congestion level
            ax3.plot(timestamps, congestion_levels, 'r-', linewidth=2)
            ax3.axhline(y=1.0, color='orange', linestyle='--', label='Congestion Threshold')
            ax3.set_title('Congestion Level')
            ax3.set_xlabel('Time (seconds)')
            ax3.set_ylabel('Congestion Factor')
            ax3.legend()
            ax3.grid(True)

            # Signal timings
            ax4.plot(timestamps, ns_green_times, 'b-', linewidth=2, label='North-South')
            ax4.plot(timestamps, ew_green_times, 'r-', linewidth=2, label='East-West')
            ax4.set_title('Adaptive Signal Timings')
            ax4.set_xlabel('Time (seconds)')
            ax4.set_ylabel('Green Time (seconds)')
            ax4.legend()
            ax4.grid(True)

            plt.tight_layout()
            plt.show()

            # Print current status
            print(f"⏱️ Time: {current_time:.1f}s | 🚗 Vehicles: {metrics['current_vehicles']}")
            print(f"🚦 Flow: {metrics['flow_rate']:.1f} v/min | 🔴 Congestion: {metrics['congestion_level']:.2f}")
            print(f"⚡ Timings - NS: {timings['north_south_green']}s, EW: {timings['east_west_green']}s")

        time.sleep(1)  # Wait 1 second

    print("\n✅ Simulation completed!")
    return timestamps, vehicle_counts, flow_rates, congestion_levels

# Run simulation
# Uncomment to run the real-time simulation
# simulate_realtime_processing()

# =============================================================================
# EXAMPLE 3: BATCH PROCESSING WITH WEBCAM OR VIDEO FILE
# =============================================================================

def process_video_file():
    """Process a video file for traffic analysis"""

    from google.colab import files
    import cv2

    print("Upload a traffic video file (MP4, AVI, MOV):")
    uploaded = files.upload()

    for filename, data in uploaded.items():
        # Save uploaded file
        with open(filename, 'wb') as f:
            f.write(data)

        # Process video
        cap = cv2.VideoCapture(filename)
        system = TrafficManagementSystem()

        frame_count = 0
        results = []

        print(f"Processing video: {filename}")

        while True:
            ret, frame = cap.read()
            if not ret:
                break

            # Process every 10th frame to speed up demo
            if frame_count % 10 == 0:
                processed_frame, metrics, timings = system.process_frame(frame)

                results.append({
                    'frame': frame_count,
                    'vehicles': metrics['current_vehicles'],
                    'flow_rate': metrics['flow_rate'],
                    'congestion': metrics['is_congested'],
                    'ns_green': timings['north_south_green'],
                    'ew_green': timings['east_west_green']
                })

                # Show progress
                if frame_count % 100 == 0:
                    print(f"Processed frame {frame_count}")

            frame_count += 1

            # Limit processing for demo (first 500 frames)
            if frame_count > 500:
                break

        cap.release()

        # Analyze results
        df = pd.DataFrame(results)

        # Plot analysis
        fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 10))

        ax1.plot(df['frame'], df['vehicles'])
        ax1.set_title('Vehicle Count Throughout Video')
        ax1.set_xlabel('Frame Number')
        ax1.set_ylabel('Number of Vehicles')

        ax2.plot(df['frame'], df['flow_rate'])
        ax2.set_title('Traffic Flow Rate')
        ax2.set_xlabel('Frame Number')
        ax2.set_ylabel('Vehicles per Minute')

        # Congestion detection
        congested_frames = df[df['congestion'] == True]['frame']
        ax3.scatter(congested_frames, [1]*len(congested_frames), color='red', alpha=0.6)
        ax3.scatter(df[df['congestion'] == False]['frame'], [0]*sum(df['congestion'] == False), color='green', alpha=0.6)
        ax3.set_title('Congestion Detection')
        ax3.set_xlabel('Frame Number')
        ax3.set_ylabel('Congested (1) / Normal (0)')

        ax4.plot(df['frame'], df['ns_green'], label='North-South', alpha=0.7)
        ax4.plot(df['frame'], df['ew_green'], label='East-West', alpha=0.7)
        ax4.set_title('Adaptive Signal Timings')
        ax4.set_xlabel('Frame Number')
        ax4.set_ylabel('Green Time (seconds)')
        ax4.legend()

        plt.tight_layout()
        plt.show()

        # Summary statistics
        print("\n📊 VIDEO ANALYSIS SUMMARY:")
        print(f"📹 Total frames processed: {len(results)}")
        print(f"🚗 Average vehicles: {df['vehicles'].mean():.1f}")
        print(f"🚗 Max vehicles: {df['vehicles'].max()}")
        print(f"🚦 Average flow rate: {df['flow_rate'].mean():.1f} vehicles/min")
        print(f"🔴 Congested frames: {sum(df['congestion'])} ({sum(df['congestion'])/len(df)*100:.1f}%)")
        print(f"⚡ Avg NS green time: {df['ns_green'].mean():.1f} seconds")
        print(f"⚡ Avg EW green time: {df['ew_green'].mean():.1f} seconds")

        break

# =============================================================================
# EXAMPLE 4: DATABASE ANALYSIS AND REPORTING
# =============================================================================

def analyze_stored_data(system):
    """Analyze historical traffic data from database"""

    # Query data from database
    cursor = system.conn.cursor()
    cursor.execute('''
        SELECT timestamp, vehicle_count, flow_rate, congestion_level, signal_timing
        FROM traffic_data
        ORDER BY timestamp
    ''')

    data = cursor.fetchall()

    if not data:
        print("No data found in database. Process some traffic first!")
        return

    # Convert to DataFrame
    df = pd.DataFrame(data, columns=['timestamp', 'vehicle_count', 'flow_rate',
                                   'congestion_level', 'signal_timing'])

    # Parse signal timing JSON
    df['signal_timing'] = df['signal_timing'].apply(json.loads)
    df['ns_green'] = df['signal_timing'].apply(lambda x: x['north_south_green'])
    df['ew_green'] = df['signal_timing'].apply(lambda x: x['east_west_green'])
    df['cycle_length'] = df['signal_timing'].apply(lambda x: x['cycle_length'])

    # Create comprehensive report
    fig, axes = plt.subplots(3, 2, figsize=(15, 12))

    # Vehicle count histogram
    axes[0,0].hist(df['vehicle_count'], bins=20, alpha=0.7, color='blue')
    axes[0,0].set_title('Vehicle Count Distribution')
    axes[0,0].set_xlabel('Number of Vehicles')
    axes[0,0].set_ylabel('Frequency')

    # Flow rate over time
    axes[0,1].plot(range(len(df)), df['flow_rate'], color='green')
    axes[0,1].set_title('Flow Rate Over Time')
    axes[0,1].set_xlabel('Observation Number')
    axes[0,1].set_ylabel('Vehicles per Minute')

    # Congestion analysis
    axes[1,0].scatter(df['vehicle_count'], df['congestion_level'], alpha=0.6)
    axes[1,0].set_title('Vehicle Count vs Congestion Level')
    axes[1,0].set_xlabel('Vehicle Count')
    axes[1,0].set_ylabel('Congestion Level')

    # Signal timing correlation
    axes[1,1].scatter(df['flow_rate'], df['cycle_length'], alpha=0.6, color='orange')
    axes[1,1].set_title('Flow Rate vs Cycle Length')
    axes[1,1].set_xlabel('Flow Rate (vehicles/min)')
    axes[1,1].set_ylabel('Cycle Length (seconds)')

    # Green time comparison
    axes[2,0].boxplot([df['ns_green'], df['ew_green']], labels=['North-South', 'East-West'])
    axes[2,0].set_title('Green Time Distribution by Direction')
    axes[2,0].set_ylabel('Green Time (seconds)')

    # System efficiency over time
    df['efficiency'] = 100 - df['congestion_level'] * 50  # Simple efficiency metric
    axes[2,1].plot(range(len(df)), df['efficiency'], color='purple')
    axes[2,1].set_title('System Efficiency Over Time')
    axes[2,1].set_xlabel('Observation Number')
    axes[2,1].set_ylabel('Efficiency Score')

    plt.tight_layout()
    plt.show()

    # Print statistics
    print("\n📈 TRAFFIC DATA ANALYSIS REPORT")
    print("=" * 50)
    print(f"📊 Total observations: {len(df)}")
    print(f"🚗 Vehicle count - Mean: {df['vehicle_count'].mean():.1f}, Std: {df['vehicle_count'].std():.1f}")
    print(f"🚦 Flow rate - Mean: {df['flow_rate'].mean():.1f}, Max: {df['flow_rate'].max():.1f}")
    print(f"🔴 Congestion level - Mean: {df['congestion_level'].mean():.2f}")
    print(f"⚡ Avg cycle length: {df['cycle_length'].mean():.1f} seconds")
    print(f"🟢 NS green time - Mean: {df['ns_green'].mean():.1f}s")
    print(f"🟢 EW green time - Mean: {df['ew_green'].mean():.1f}s")

    # Correlation analysis
    correlation_matrix = df[['vehicle_count', 'flow_rate', 'congestion_level',
                           'ns_green', 'ew_green', 'cycle_length']].corr()

    plt.figure(figsize=(10, 8))
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0)
    plt.title('Traffic Metrics Correlation Matrix')
    plt.tight_layout()
    plt.show()

# =============================================================================
# EXAMPLE 5: PERFORMANCE BENCHMARKING
# =============================================================================

def benchmark_system_performance():
    """Benchmark different components of the system"""

    import time

    print("🔬 SYSTEM PERFORMANCE BENCHMARKING")
    print("=" * 50)

    # Create test data
    system = TrafficManagementSystem()

    # Generate synthetic frame for testing
    test_frame = np.random.randint(0, 255, (720, 1280, 3), dtype=np.uint8)

    # Benchmark vehicle detection
    print("\n🚗 Vehicle Detection Performance:")
    detection_times = []
    for i in range(10):
        start_time = time.time()
        detections = system.detector.detect_vehicles(test_frame)
        detection_times.append(time.time() - start_time)

    avg_detection_time = np.mean(detection_times)
    fps_detection = 1 / avg_detection_time
    print(f"   Average detection time: {avg_detection_time:.3f} seconds")
    print(f"   Detection FPS: {fps_detection:.1f}")
    print(f"   Detections found: {len(detections)}")

    # Benchmark tracking
    print("\n🎯 Vehicle Tracking Performance:")
    tracking_times = []
    for i in range(100):
        start_time = time.time()
        tracked = system.tracker.update(detections)
        tracking_times.append(time.time() - start_time)

    avg_tracking_time = np.mean(tracking_times)
    print(f"   Average tracking time: {avg_tracking_time:.4f} seconds")
    print(f"   Tracking objects: {len(tracked)}")

    # Benchmark full pipeline
    print("\n⚡ Full Pipeline Performance:")
    pipeline_times = []
    for i in range(5):
        start_time = time.time()
        processed_frame, metrics, timings = system.process_frame(test_frame)
        pipeline_times.append(time.time() - start_time)

    avg_pipeline_time = np.mean(pipeline_times)
    fps_pipeline = 1 / avg_pipeline_time
    print(f"   Average pipeline time: {avg_pipeline_time:.3f} seconds")
    print(f"   Pipeline FPS: {fps_pipeline:.1f}")

    # Memory usage estimation
    import sys
    frame_size_mb = test_frame.nbytes / (1024 * 1024)
    print(f"\n💾 Memory Usage:")
    print(f"   Frame size: {frame_size_mb:.1f} MB")
    print(f"   Estimated memory for 30 FPS: {frame_size_mb * 30:.1f} MB/second")

    # Performance recommendations
    print(f"\n📋 Performance Assessment:")
    if fps_pipeline >= 15:
        print("   ✅ System can handle real-time processing")
    elif fps_pipeline >= 5:
        print("   ⚠️ System suitable for reduced frame rate processing")
    else:
        print("   ❌ System needs optimization for real-time use")

    print(f"   💡 Recommendations:")
    if avg_detection_time > 0.1:
        print("     - Consider using a smaller YOLO model (yolov8n vs yolov8s)")
        print("     - Reduce input frame resolution")
    if len(system.tracker.objects) > 100:
        print("     - Implement object cleanup for disappeared vehicles")

    return {
        'detection_fps': fps_detection,
        'pipeline_fps': fps_pipeline,
        'avg_detection_time': avg_detection_time,
        'avg_pipeline_time': avg_pipeline_time
    }

# =============================================================================
# EXAMPLE 6: ADVANCED FEATURES DEMONSTRATION
# =============================================================================

def demonstrate_advanced_features():
    """Demonstrate advanced system capabilities"""

    print("🚀 ADVANCED FEATURES DEMONSTRATION")
    print("=" * 50)

    # Multi-intersection simulation
    print("\n🏙️ Multi-Intersection Coordination:")

    # Create multiple intersection systems
    intersections = {
        'Main_1st': TrafficManagementSystem(),
        'Main_2nd': TrafficManagementSystem(),
        'Oak_1st': TrafficManagementSystem()
    }

    # Simulate coordinated traffic flow
    for i in range(20):
        print(f"\n⏰ Time Step {i+1}:")

        for intersection_name, system in intersections.items():
            # Simulate varying traffic at different intersections
            base_traffic = 10 + 5 * np.sin(i * 0.3) + random.randint(-3, 3)
            if intersection_name == 'Main_1st':
                base_traffic += 5  # Main street has more traffic

            system.analyzer.update_vehicle_count(max(0, int(base_traffic)))
            metrics = system.analyzer.get_traffic_metrics()

            traffic_data = {
                'north_south_flow': metrics['flow_rate'] * 0.6,
                'east_west_flow': metrics['flow_rate'] * 0.4,
                'congestion_level': metrics['congestion_level']
            }

            timings = system.optimizer.calculate_optimal_timing(traffic_data)

            print(f"   {intersection_name}: {metrics['current_vehicles']} vehicles, "
                  f"Cycle: {timings['cycle_length']}s")

    # Traffic prediction demonstration
    print(f"\n🔮 Traffic Prediction Capabilities:")

    # Simple trend analysis
    historical_counts = [15, 18, 22, 19, 25, 30, 28, 24, 20, 16]

    # Linear regression for prediction
    from sklearn.linear_model import LinearRegression
    X = np.array(range(len(historical_counts))).reshape(-1, 1)
    y = np.array(historical_counts)

    model = LinearRegression()
    model.fit(X, y)

    # Predict next 5 time steps
    future_X = np.array(range(len(historical_counts), len(historical_counts) + 5)).reshape(-1, 1)
    predictions = model.predict(future_X)

    plt.figure(figsize=(12, 6))
    plt.plot(range(len(historical_counts)), historical_counts, 'bo-', label='Historical Data')
    plt.plot(range(len(historical_counts), len(historical_counts) + 5), predictions, 'ro--', label='Predictions')
    plt.title('Traffic Prediction Example')
    plt.xlabel('Time Steps')
    plt.ylabel('Vehicle Count')
    plt.legend()
    plt.grid(True)
    plt.show()

    print(f"   Predicted vehicle counts: {[int(p) for p in predictions]}")

    # Emergency vehicle priority simulation
    print(f"\n🚨 Emergency Vehicle Priority System:")

    def emergency_override(system, direction='north_south'):
        """Simulate emergency vehicle priority"""
        emergency_timings = {
            'north_south_green': 60 if direction == 'north_south' else 15,
            'east_west_green': 15 if direction == 'north_south' else 60,
            'cycle_length': 80,
            'yellow_time': 3,
            'red_clearance': 2
        }

        print(f"   Emergency vehicle detected on {direction.replace('_', '-')} route")
        print(f"   Override timings: NS={emergency_timings['north_south_green']}s, "
              f"EW={emergency_timings['east_west_green']}s")

        return emergency_timings

    # Simulate emergency scenario
    system = TrafficManagementSystem()
    normal_timings = system.optimizer.calculate_optimal_timing({
        'north_south_flow': 20,
        'east_west_flow': 15,
        'congestion_level': 1.0
    })

    emergency_timings = emergency_override(system, 'north_south')

    print(f"   Normal: NS={normal_timings['north_south_green']}s, EW={normal_timings['east_west_green']}s")
    print(f"   Emergency: NS={emergency_timings['north_south_green']}s, EW={emergency_timings['east_west_green']}s")

# =============================================================================
# INTERACTIVE TESTING FUNCTIONS
# =============================================================================

def create_interactive_test_suite():
    """Create an interactive test suite with buttons"""

    # Create buttons for different tests
    test_detection_btn = widgets.Button(description="Test Vehicle Detection", button_style='info')
    test_simulation_btn = widgets.Button(description="Run Traffic Simulation", button_style='success')
    test_video_btn = widgets.Button(description="Process Video File", button_style='warning')
    benchmark_btn = widgets.Button(description="Run Benchmarks", button_style='danger')
    advanced_demo_btn = widgets.Button(description="Advanced Features", button_style='')

    output = widgets.Output()

    # Button click handlers
    def on_detection_test(b):
        with output:
            clear_output()
            test_vehicle_detection()

    def on_simulation_test(b):
        with output:
            clear_output()
            simulate_realtime_processing()

    def on_video_test(b):
        with output:
            clear_output()
            process_video_file()

    def on_benchmark_test(b):
        with output:
            clear_output()
            benchmark_system_performance()

    def on_advanced_demo(b):
        with output:
            clear_output()
            demonstrate_advanced_features()

    # Connect handlers
    test_detection_btn.on_click(on_detection_test)
    test_simulation_btn.on_click(on_simulation_test)
    test_video_btn.on_click(on_video_test)
    benchmark_btn.on_click(on_benchmark_test)
    advanced_demo_btn.on_click(on_advanced_demo)

    # Layout
    button_row1 = widgets.HBox([test_detection_btn, test_simulation_btn, test_video_btn])
    button_row2 = widgets.HBox([benchmark_btn, advanced_demo_btn])

    test_suite = widgets.VBox([
        widgets.HTML("<h2>🧪 Traffic Management System Test Suite</h2>"),
        button_row1,
        button_row2,
        output
    ])

    return test_suite

# =============================================================================
# MAIN EXECUTION FOR TESTING
# =============================================================================

print("🧪 Traffic Management System - Testing & Examples Loaded!")
print("\nAvailable test functions:")
print("- process_uploaded_image(): Test with your own traffic images")
print("- simulate_realtime_processing(): Real-time simulation with synthetic data")
print("- process_video_file(): Analyze uploaded traffic videos")
print("- benchmark_system_performance(): Performance testing")
print("- demonstrate_advanced_features(): Advanced capabilities demo")

# Create and display interactive test suite
interactive_suite = create_interactive_test_suite()
display(interactive_suite)

# Quick start example
print("\n🚀 Quick Start Example:")
print("Run this to test the system with a simple example:")
print()
print("# Initialize and test system")
print("system = TrafficManagementSystem()")
print("print('System ready for testing!')")
```

```python
# TESTING AND DEMO FUNCTIONS

def test_vehicle_detection():
    """Test vehicle detection with sample image"""
    print("Testing vehicle detection...")

    # Create a sample traffic scene (you can upload your own image)
    detector = VehicleDetector()

    # For demo, we'll create a simple test
    # In practice, you'd load a real traffic image
    print("Vehicle detection system ready!")
    print("Upload a traffic image to test detection capabilities.")

    return detector

def create_dashboard_widgets():
    """Create interactive dashboard widgets"""

    # Control widgets
    start_button = widgets.Button(description="Start System")
    stop_button = widgets.Button(description="Stop System")
    upload_button = widgets.FileUpload(accept='image/*,video/*', multiple=False)

    # Display widgets
    output = widgets.Output()

    # Metrics display
    metrics_html = widgets.HTML(value="<h3>System Metrics</h3><p>System not started</p>")

    # Layout
    controls = widgets.HBox([start_button, stop_button, upload_button])
    dashboard = widgets.VBox([controls, metrics_html, output])

    return dashboard, start_button, stop_button, upload_button, output, metrics_html

def run_system_demo():
    """Run a complete system demonstration"""
    print("=" * 60)
    print("TRAFFIC MANAGEMENT SYSTEM DEMO")
    print("=" * 60)

    # Initialize system with a pre-trained model
    # Using a pre-trained model like 'yolov8n.pt' for demonstration
    system = TrafficManagementSystem(model_path='yolov8n.pt') # Pass model_path instead of model_name

    print("\n✅ System initialized successfully!")
    print("\nSystem Components:")
    print("- Vehicle Detection: YOLO-based object detection (using pre-trained model)")
    print("- Vehicle Tracking: Centroid-based tracking")
    print("- Traffic Analysis: Flow rate and congestion detection")
    print("- Signal Optimization: Adaptive timing algorithms")
    print("- Data Storage: SQLite database")

    print("\n📊 System ready for processing!")
    print("Upload a traffic video or image to see the system in action.")

    return system

def handle_upload(change):
    """Handle file upload and trigger processing"""
    if not system.is_running:
        with output:
            clear_output()
            print("System is not running. Please click 'Start System' first.")
        return

    uploaded_file = upload_btn.value
    if not uploaded_file:
        with output:
            clear_output()
            print("No file uploaded.")
        return

    filename = list(uploaded_file.keys())[0]
    content = uploaded_file[filename]['content']

    with output:
        clear_output(wait=True)
        print(f"Processing file: {filename}")

        # Check if the uploaded file is a video (basic check based on extension)
        if filename.lower().endswith(('.mp4', '.avi', '.mov', '.mkv')):
            # Handle video processing

            # Save the uploaded video content to a temporary file
            temp_video_path = f"/tmp/{filename}"
            with open(temp_video_path, 'wb') as f:
                f.write(content)

            cap = cv2.VideoCapture(temp_video_path)

            if not cap.isOpened():
                print(f"Error: Could not open video file {filename}.")
                return

            # Get video properties
            fps = int(cap.get(cv2.CAP_PROP_FPS))
            frame_count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
            duration = frame_count / fps if fps > 0 else 0

            print(f"Video properties: {frame_count} frames, {fps} fps, {duration:.2f} seconds duration")

            # Create a widget to display video frames
            image_widget = widgets.Image(format='jpeg', width=600, height=400)
            display(image_widget)


            frame_num = 0
            while cap.isOpened():
                ret, frame = cap.read()
                if not ret:
                    break

                frame_num += 1
                # Process the frame
                annotated_frame, traffic_metrics, optimal_timings = system.process_frame(frame)

                # Update metrics display for the current frame
                metrics.value = f"""
                <h3>System Metrics (Frame {frame_num}/{frame_count})</h3>
                <p><b>Vehicles Detected:</b> {traffic_metrics['current_vehicles']}</p>
                <p><b>Flow Rate:</b> {traffic_metrics['flow_rate']:.1f} veh/min</p>
                <p><b>Congestion Level:</b> {traffic_metrics['congestion_level']:.2f}</p>
                <p><b>Is Congested:</b> {'YES' if traffic_metrics['is_congested'] else 'NO'}</p>
                <h4>Optimal Signal Timings:</h4>
                <p><b>North-South Green:</b> {optimal_timings['north_south_green']}s</p>
                <p><b>East-West Green:</b> {optimal_timings['east_west_green']}s</p>
                <p><b>Cycle Length:</b> {optimal_timings['cycle_length']}s</p>
                """
                # Convert the frame to JPEG format for display
                _, buffer = cv2.imencode('.jpg', annotated_frame)
                image_widget.value = buffer.tobytes()

                # Optional: Add a small delay to control playback speed
                # time.sleep(1 / fps)

            cap.release()
            print(f"Finished processing video: {filename}")

        else:
            # Handle image processing (existing logic)
            image_array = cv2.imdecode(np.frombuffer(content, np.uint8), cv2.IMREAD_COLOR)

            if image_array is None:
                print(f"Error: Could not decode image from file '{filename}'. Please check the file format.")
                return

            frame = image_array

            # Process the frame
            annotated_frame, traffic_metrics, optimal_timings = system.process_frame(frame)

            # Display the annotated frame
            plt.imshow(cv2.cvtColor(annotated_frame, cv2.COLOR_BGR2RGB))
            plt.axis('off')
            plt.show()

            # Update metrics display
            metrics.value = f"""
            <h3>System Metrics</h3>
            <p><b>Vehicles Detected:</b> {traffic_metrics['current_vehicles']}</p>
            <p><b>Flow Rate:</b> {traffic_metrics['flow_rate']:.1f} veh/min</p>
            <p><b>Congestion Level:</b> {traffic_metrics['congestion_level']:.2f}</p>
            <p><b>Is Congested:</b> {'YES' if traffic_metrics['is_congested'] else 'NO'}</p>
            <h4>Optimal Signal Timings:</h4>
            <p><b>North-South Green:</b> {optimal_timings['north_south_green']}s</p>
            <p><b>East-West Green:</b> {optimal_timings['east_west_green']}s</p>
            <p><b>Cycle Length:</b> {optimal_timings['cycle_length']}s</p>
            """


def on_start_button_clicked(b):
    """Handle Start System button click"""
    with output:
        clear_output()
        if not system.is_running:
            system.is_running = True
            print("Traffic Management System started. Upload an image or video.")
            metrics.value = "<h3>System Metrics</h3><p>System is running. Waiting for input.</p>"
        else:
            print("System is already running.")

def on_stop_button_clicked(b):
    """Handle Stop System button click"""
    with output:
        clear_output()
        if system.is_running:
            system.is_running = False
            print("Traffic Management System stopped.")
            metrics.value = "<h3>System Metrics</h3><p>System stopped.</p>"
        else:
            print("System is already stopped.")

# =============================================================================
# MAIN EXECUTION
# =============================================================================

if __name__ == "__main__":
    # Run the demo
    system = run_system_demo()

    # Create interactive dashboard
    dashboard, start_btn, stop_btn, upload_btn, output, metrics = create_dashboard_widgets()
    display(dashboard)

    # Link button clicks to handlers
    start_btn.on_click(on_start_button_clicked)
    stop_btn.on_click(on_stop_button_clicked)
    upload_btn.observe(handle_upload, names='value')

    print("Event handlers linked to dashboard widgets.")
    print("Click 'Start System' and upload an image or video to test.")

# =============================================================================
# EXAMPLE 1: PROCESS UPLOADED IMAGE
# =============================================================================

def process_uploaded_image():
    """Process an uploaded traffic image"""

    # Create file upload widget
    from google.colab import files
    import cv2
    import numpy as np
    from PIL import Image
    import matplotlib.pyplot as plt

    print("Upload a traffic image (JPG, PNG):")
    uploaded = files.upload()

    for filename, data in uploaded.items():
        # Load image
        image = Image.open(io.BytesIO(data))
        frame = cv2.cvtColor(np.array(image), cv2.COLOR_RGB2BGR)

        # Initialize system and process
        system = TrafficManagementSystem()
        processed_frame, metrics, timings = system.process_frame(frame)

        # Display results
        fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

        # Original image
        ax1.imshow(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
        ax1.set_title("Original Image")
        ax1.axis('off')

        # Processed image with detections
        ax2.imshow(cv2.cvtColor(processed_frame, cv2.COLOR_BGR2RGB))
        ax2.set_title("Vehicle Detection & Analysis")
        ax2.axis('off')

        plt.tight_layout()
        plt.show()

        # Print metrics
        print("\n📊 TRAFFIC ANALYSIS RESULTS:")
        print(f"🚗 Vehicles Detected: {metrics['current_vehicles']}")
        print(f"🚦 Flow Rate: {metrics['flow_rate']:.1f} vehicles/minute")
        print(f"🔴 Congestion Status: {'CONGESTED' if metrics['is_congested'] else 'NORMAL'}")
        print(f"📈 Congestion Level: {metrics['congestion_level']:.2f}")

        print("\n⚡ OPTIMIZED SIGNAL TIMINGS:")
        print(f"🟢 North-South Green: {timings['north_south_green']} seconds")
        print(f"🟢 East-West Green: {timings['east_west_green']} seconds")
        print(f"🔄 Total Cycle Length: {timings['cycle_length']} seconds")

        break

# Run the example
# Uncomment the line below to test with an image upload
# process_uploaded_image()

# =============================================================================
# EXAMPLE 2: SIMULATE REAL-TIME PROCESSING WITH SAMPLE DATA
# =============================================================================

def simulate_realtime_processing():
    """Simulate real-time traffic processing with synthetic data"""

    import time
    import random
    from IPython.display import clear_output
    import matplotlib.pyplot as plt

    print("🚀 Starting Real-Time Traffic Simulation...")
    print("This will simulate 60 seconds of traffic processing\n")

    # Initialize system
    system = TrafficManagementSystem()

    # Storage for plotting
    timestamps = []
    vehicle_counts = []
    flow_rates = []
    congestion_levels = []
    ns_green_times = []
    ew_green_times = []

    start_time = time.time()

    for i in range(60):  # 60 seconds simulation
        current_time = time.time() - start_time

        # Simulate varying traffic conditions
        base_vehicles = 15 + 10 * np.sin(current_time * 0.1) + random.randint(-5, 5)
        base_vehicles = max(0, int(base_vehicles))

        # Create mock traffic data
        system.analyzer.update_vehicle_count(base_vehicles)
        metrics = system.analyzer.get_traffic_metrics()

        # Get optimal timings
        traffic_data = {
            'north_south_flow': metrics['flow_rate'] * (0.4 + 0.2 * np.sin(current_time * 0.05)),
            'east_west_flow': metrics['flow_rate'] * (0.6 - 0.2 * np.sin(current_time * 0.05)),
            'congestion_level': metrics['congestion_level']
        }
        timings = system.optimizer.calculate_optimal_timing(traffic_data)

        # Store data
        timestamps.append(current_time)
        vehicle_counts.append(metrics['current_vehicles'])
        flow_rates.append(metrics['flow_rate'])
        congestion_levels.append(metrics['congestion_level'])
        ns_green_times.append(timings['north_south_green'])
        ew_green_times.append(timings['east_west_green'])

        # Update display every 5 seconds
        if i % 5 == 0:
            clear_output(wait=True)

            fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 10))

            # Vehicle count over time
            ax1.plot(timestamps, vehicle_counts, 'b-', linewidth=2)
            ax1.set_title('Vehicle Count Over Time')
            ax1.set_xlabel('Time (seconds)')
            ax1.set_ylabel('Number of Vehicles')
            ax1.grid(True)

            # Flow rate
            ax2.plot(timestamps, flow_rates, 'g-', linewidth=2)
            ax2.set_title('Traffic Flow Rate')
            ax2.set_xlabel('Time (seconds)')
            ax2.set_ylabel('Vehicles per Minute')
            ax2.grid(True)

            # Congestion level
            ax3.plot(timestamps, congestion_levels, 'r-', linewidth=2)
            ax3.axhline(y=1.0, color='orange', linestyle='--', label='Congestion Threshold')
            ax3.set_title('Congestion Level')
            ax3.set_xlabel('Time (seconds)')
            ax3.set_ylabel('Congestion Factor')
            ax3.legend()
            ax3.grid(True)

            # Signal timings
            ax4.plot(timestamps, ns_green_times, 'b-', linewidth=2, label='North-South')
            ax4.plot(timestamps, ew_green_times, 'r-', linewidth=2, label='East-West')
            ax4.set_title('Adaptive Signal Timings')
            ax4.set_xlabel('Time (seconds)')
            ax4.set_ylabel('Green Time (seconds)')
            ax4.legend()
            ax4.grid(True)

            plt.tight_layout()
            plt.show()

            # Print current status
            print(f"⏱️ Time: {current_time:.1f}s | 🚗 Vehicles: {metrics['current_vehicles']}")
            print(f"🚦 Flow: {metrics['flow_rate']:.1f} v/min | 🔴 Congestion: {metrics['congestion_level']:.2f}")
            print(f"⚡ Timings - NS: {timings['north_south_green']}s, EW: {timings['east_west_green']}s")

        time.sleep(1)  # Wait 1 second

    print("\n✅ Simulation completed!")
    return timestamps, vehicle_counts, flow_rates, congestion_levels

# Run simulation
# Uncomment to run the real-time simulation
# simulate_realtime_processing()

# =============================================================================
# EXAMPLE 3: BATCH PROCESSING WITH WEBCAM OR VIDEO FILE
# =============================================================================

def process_video_file():
    """Process a video file for traffic analysis"""

    from google.colab import files
    import cv2

    print("Upload a traffic video file (MP4, AVI, MOV):")
    uploaded = files.upload()

    for filename, data in uploaded.items():
        # Save uploaded file
        with open(filename, 'wb') as f:
            f.write(data)

        # Process video
        cap = cv2.VideoCapture(filename)
        system = TrafficManagementSystem()

        frame_count = 0
        results = []

        print(f"Processing video: {filename}")

        while True:
            ret, frame = cap.read()
            if not ret:
                break

            # Process every 10th frame to speed up demo
            if frame_count % 10 == 0:
                processed_frame, metrics, timings = system.process_frame(frame)

                results.append({
                    'frame': frame_count,
                    'vehicles': metrics['current_vehicles'],
                    'flow_rate': metrics['flow_rate'],
                    'congestion': metrics['is_congested'],
                    'ns_green': timings['north_south_green'],
                    'ew_green': timings['east_west_green']
                })

                # Show progress
                if frame_count % 100 == 0:
                    print(f"Processed frame {frame_count}")

            frame_count += 1

            # Limit processing for demo (first 500 frames)
            if frame_count > 500:
                break

        cap.release()

        # Analyze results
        df = pd.DataFrame(results)

        # Plot analysis
        fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 10))

        ax1.plot(df['frame'], df['vehicles'])
        ax1.set_title('Vehicle Count Throughout Video')
        ax1.set_xlabel('Frame Number')
        ax1.set_ylabel('Number of Vehicles')

        ax2.plot(df['frame'], df['flow_rate'])
        ax2.set_title('Traffic Flow Rate')
        ax2.set_xlabel('Frame Number')
        ax2.set_ylabel('Vehicles per Minute')

        # Congestion detection
        congested_frames = df[df['congestion'] == True]['frame']
        ax3.scatter(congested_frames, [1]*len(congested_frames), color='red', alpha=0.6)
        ax3.scatter(df[df['congestion'] == False]['frame'], [0]*sum(df['congestion'] == False), color='green', alpha=0.6)
        ax3.set_title('Congestion Detection')
        ax3.set_xlabel('Frame Number')
        ax3.set_ylabel('Congested (1) / Normal (0)')

        ax4.plot(df['frame'], df['ns_green'], label='North-South', alpha=0.7)
        ax4.plot(df['frame'], df['ew_green'], label='East-West', alpha=0.7)
        ax4.set_title('Adaptive Signal Timings')
        ax4.set_xlabel('Frame Number')
        ax4.set_ylabel('Green Time (seconds)')
        ax4.legend()

        plt.tight_layout()
        plt.show()

        # Summary statistics
        print("\n📊 VIDEO ANALYSIS SUMMARY:")
        print(f"📹 Total frames processed: {len(results)}")
        print(f"🚗 Average vehicles: {df['vehicles'].mean():.1f}")
        print(f"🚗 Max vehicles: {df['vehicles'].max()}")
        print(f"🚦 Average flow rate: {df['flow_rate'].mean():.1f} vehicles/min")
        print(f"🔴 Congested frames: {sum(df['congestion'])} ({sum(df['congestion'])/len(df)*100:.1f}%)")
        print(f"⚡ Avg NS green time: {df['ns_green'].mean():.1f} seconds")
        print(f"⚡ Avg EW green time: {df['ew_green'].mean():.1f} seconds")

        break

# =============================================================================
# EXAMPLE 4: DATABASE ANALYSIS AND REPORTING
# =============================================================================

def analyze_stored_data(system):
    """Analyze historical traffic data from database"""

    # Query data from database
    cursor = system.conn.cursor()
    cursor.execute('''
        SELECT timestamp, vehicle_count, flow_rate, congestion_level, signal_timing
        FROM traffic_data
        ORDER BY timestamp
    ''')

    data = cursor.fetchall()

    if not data:
        print("No data found in database. Process some traffic first!")
        return

    # Convert to DataFrame
    df = pd.DataFrame(data, columns=['timestamp', 'vehicle_count', 'flow_rate',
                                   'congestion_level', 'signal_timing'])

    # Parse signal timing JSON
    df['signal_timing'] = df['signal_timing'].apply(json.loads)
    df['ns_green'] = df['signal_timing'].apply(lambda x: x['north_south_green'])
    df['ew_green'] = df['signal_timing'].apply(lambda x: x['east_west_green'])
    df['cycle_length'] = df['signal_timing'].apply(lambda x: x['cycle_length'])

    # Create comprehensive report
    fig, axes = plt.subplots(3, 2, figsize=(15, 12))

    # Vehicle count histogram
    axes[0,0].hist(df['vehicle_count'], bins=20, alpha=0.7, color='blue')
    axes[0,0].set_title('Vehicle Count Distribution')
    axes[0,0].set_xlabel('Number of Vehicles')
    axes[0,0].set_ylabel('Frequency')

    # Flow rate over time
    axes[0,1].plot(range(len(df)), df['flow_rate'], color='green')
    axes[0,1].set_title('Flow Rate Over Time')
    axes[0,1].set_xlabel('Observation Number')
    axes[0,1].set_ylabel('Vehicles per Minute')

    # Congestion analysis
    axes[1,0].scatter(df['vehicle_count'], df['congestion_level'], alpha=0.6)
    axes[1,0].set_title('Vehicle Count vs Congestion Level')
    axes[1,0].set_xlabel('Vehicle Count')
    axes[1,0].set_ylabel('Congestion Level')

    # Signal timing correlation
    axes[1,1].scatter(df['flow_rate'], df['cycle_length'], alpha=0.6, color='orange')
    axes[1,1].set_title('Flow Rate vs Cycle Length')
    axes[1,1].set_xlabel('Flow Rate (vehicles/min)')
    axes[1,1].set_ylabel('Cycle Length (seconds)')

    # Green time comparison
    axes[2,0].boxplot([df['ns_green'], df['ew_green']], labels=['North-South', 'East-West'])
    axes[2,0].set_title('Green Time Distribution by Direction')
    axes[2,0].set_ylabel('Green Time (seconds)')

    # System efficiency over time
    df['efficiency'] = 100 - df['congestion_level'] * 50  # Simple efficiency metric
    axes[2,1].plot(range(len(df)), df['efficiency'], color='purple')
    axes[2,1].set_title('System Efficiency Over Time')
    axes[2,1].set_xlabel('Observation Number')
    axes[2,1].set_ylabel('Efficiency Score')

    plt.tight_layout()
    plt.show()

    # Print statistics
    print("\n📈 TRAFFIC DATA ANALYSIS REPORT")
    print("=" * 50)
    print(f"📊 Total observations: {len(df)}")
    print(f"🚗 Vehicle count - Mean: {df['vehicle_count'].mean():.1f}, Std: {df['vehicle_count'].std():.1f}")
    print(f"🚦 Flow rate - Mean: {df['flow_rate'].mean():.1f}, Max: {df['flow_rate'].max():.1f}")
    print(f"🔴 Congestion level - Mean: {df['congestion_level'].mean():.2f}")
    print(f"⚡ Avg cycle length: {df['cycle_length'].mean():.1f} seconds")
    print(f"🟢 NS green time - Mean: {df['ns_green'].mean():.1f}s")
    print(f"🟢 EW green time - Mean: {df['ew_green'].mean():.1f}s")

    # Correlation analysis
    correlation_matrix = df[['vehicle_count', 'flow_rate', 'congestion_level',
                           'ns_green', 'ew_green', 'cycle_length']].corr()

    plt.figure(figsize=(10, 8))
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0)
    plt.title('Traffic Metrics Correlation Matrix')
    plt.tight_layout()
    plt.show()

# =============================================================================
# EXAMPLE 5: PERFORMANCE BENCHMARKING
# =============================================================================

def benchmark_system_performance():
    """Benchmark different components of the system"""

    import time

    print("🔬 SYSTEM PERFORMANCE BENCHMARKING")
    print("=" * 50)

    # Create test data
    system = TrafficManagementSystem()

    # Generate synthetic frame for testing
    test_frame = np.random.randint(0, 255, (720, 1280, 3), dtype=np.uint8)

    # Benchmark vehicle detection
    print("\n🚗 Vehicle Detection Performance:")
    detection_times = []
    for i in range(10):
        start_time = time.time()
        detections = system.detector.detect_vehicles(test_frame)
        detection_times.append(time.time() - start_time)

    avg_detection_time = np.mean(detection_times)
    fps_detection = 1 / avg_detection_time
    print(f"   Average detection time: {avg_detection_time:.3f} seconds")
    print(f"   Detection FPS: {fps_detection:.1f}")
    print(f"   Detections found: {len(detections)}")

    # Benchmark tracking
    print("\n🎯 Vehicle Tracking Performance:")
    tracking_times = []
    for i in range(100):
        start_time = time.time()
        tracked = system.tracker.update(detections)
        tracking_times.append(time.time() - start_time)

    avg_tracking_time = np.mean(tracking_times)
    print(f"   Average tracking time: {avg_tracking_time:.4f} seconds")
    print(f"   Tracking objects: {len(tracked)}")

    # Benchmark full pipeline
    print("\n⚡ Full Pipeline Performance:")
    pipeline_times = []
    for i in range(5):
        start_time = time.time()
        processed_frame, metrics, timings = system.process_frame(test_frame)
        pipeline_times.append(time.time() - start_time)

    avg_pipeline_time = np.mean(pipeline_times)
    fps_pipeline = 1 / avg_pipeline_time
    print(f"   Average pipeline time: {avg_pipeline_time:.3f} seconds")
    print(f"   Pipeline FPS: {fps_pipeline:.1f}")

    # Memory usage estimation
    import sys
    frame_size_mb = test_frame.nbytes / (1024 * 1024)
    print(f"\n💾 Memory Usage:")
    print(f"   Frame size: {frame_size_mb:.1f} MB")
    print(f"   Estimated memory for 30 FPS: {frame_size_mb * 30:.1f} MB/second")

    # Performance recommendations
    print(f"\n📋 Performance Assessment:")
    if fps_pipeline >= 15:
        print("   ✅ System can handle real-time processing")
    elif fps_pipeline >= 5:
        print("   ⚠️ System suitable for reduced frame rate processing")
    else:
        print("   ❌ System needs optimization for real-time use")

    print(f"   💡 Recommendations:")
    if avg_detection_time > 0.1:
        print("     - Consider using a smaller YOLO model (yolov8n vs yolov8s)")
        print("     - Reduce input frame resolution")
    if len(system.tracker.objects) > 100:
        print("     - Implement object cleanup for disappeared vehicles")

    return {
        'detection_fps': fps_detection,
        'pipeline_fps': fps_pipeline,
        'avg_detection_time': avg_detection_time,
        'avg_pipeline_time': avg_pipeline_time
    }

# =============================================================================
# EXAMPLE 6: ADVANCED FEATURES DEMONSTRATION
# =============================================================================

def demonstrate_advanced_features():
    """Demonstrate advanced system capabilities"""

    print("🚀 ADVANCED FEATURES DEMONSTRATION")
    print("=" * 50)

    # Multi-intersection simulation
    print("\n🏙️ Multi-Intersection Coordination:")

    # Create multiple intersection systems
    intersections = {
        'Main_1st': TrafficManagementSystem(),
        'Main_2nd': TrafficManagementSystem(),
        'Oak_1st': TrafficManagementSystem()
    }

    # Simulate coordinated traffic flow
    for i in range(20):
        print(f"\n⏰ Time Step {i+1}:")

        for intersection_name, system in intersections.items():
            # Simulate varying traffic at different intersections
            base_traffic = 10 + 5 * np.sin(i * 0.3) + random.randint(-3, 3)
            if intersection_name == 'Main_1st':
                base_traffic += 5  # Main street has more traffic

            system.analyzer.update_vehicle_count(max(0, int(base_traffic)))
            metrics = system.analyzer.get_traffic_metrics()

            traffic_data = {
                'north_south_flow': metrics['flow_rate'] * 0.6,
                'east_west_flow': metrics['flow_rate'] * 0.4,
                'congestion_level': metrics['congestion_level']
            }

            timings = system.optimizer.calculate_optimal_timing(traffic_data)

            print(f"   {intersection_name}: {metrics['current_vehicles']} vehicles, "
                  f"Cycle: {timings['cycle_length']}s")

    # Traffic prediction demonstration
    print(f"\n🔮 Traffic Prediction Capabilities:")

    # Simple trend analysis
    historical_counts = [15, 18, 22, 19, 25, 30, 28, 24, 20, 16]

    # Linear regression for prediction
    from sklearn.linear_model import LinearRegression
    X = np.array(range(len(historical_counts))).reshape(-1, 1)
    y = np.array(historical_counts)

    model = LinearRegression()
    model.fit(X, y)

    # Predict next 5 time steps
    future_X = np.array(range(len(historical_counts), len(historical_counts) + 5)).reshape(-1, 1)
    predictions = model.predict(future_X)

    plt.figure(figsize=(12, 6))
    plt.plot(range(len(historical_counts)), historical_counts, 'bo-', label='Historical Data')
    plt.plot(range(len(historical_counts), len(historical_counts) + 5), predictions, 'ro--', label='Predictions')
    plt.title('Traffic Prediction Example')
    plt.xlabel('Time Steps')
    plt.ylabel('Vehicle Count')
    plt.legend()
    plt.grid(True)
    plt.show()

    print(f"   Predicted vehicle counts: {[int(p) for p in predictions]}")

    # Emergency vehicle priority simulation
    print(f"\n🚨 Emergency Vehicle Priority System:")

    def emergency_override(system, direction='north_south'):
        """Simulate emergency vehicle priority"""
        emergency_timings = {
            'north_south_green': 60 if direction == 'north_south' else 15,
            'east_west_green': 15 if direction == 'north_south' else 60,
            'cycle_length': 80,
            'yellow_time': 3,
            'red_clearance': 2
        }

        print(f"   Emergency vehicle detected on {direction.replace('_', '-')} route")
        print(f"   Override timings: NS={emergency_timings['north_south_green']}s, "
              f"EW={emergency_timings['east_west_green']}s")

        return emergency_timings

    # Simulate emergency scenario
    system = TrafficManagementSystem()
    normal_timings = system.optimizer.calculate_optimal_timing({
        'north_south_flow': 20,
        'east_west_flow': 15,
        'congestion_level': 1.0
    })

    emergency_timings = emergency_override(system, 'north_south')

    print(f"   Normal: NS={normal_timings['north_south_green']}s, EW={normal_timings['east_west_green']}s")
    print(f"   Emergency: NS={emergency_timings['north_south_green']}s, EW={emergency_timings['east_west_green']}s")

# =============================================================================
# INTERACTIVE TESTING FUNCTIONS
# =============================================================================

def create_interactive_test_suite():
    """Create an interactive test suite with buttons"""

    # Create buttons for different tests
    test_detection_btn = widgets.Button(description="Test Vehicle Detection", button_style='info')
    test_simulation_btn = widgets.Button(description="Run Traffic Simulation", button_style='success')
    test_video_btn = widgets.Button(description="Process Video File", button_style='warning')
    benchmark_btn = widgets.Button(description="Run Benchmarks", button_style='danger')
    advanced_demo_btn = widgets.Button(description="Advanced Features", button_style='')

    output = widgets.Output()

    # Button click handlers
    def on_detection_test(b):
        with output:
            clear_output()
            test_vehicle_detection()

    def on_simulation_test(b):
        with output:
            clear_output()
            simulate_realtime_processing()

    def on_video_test(b):
        with output:
            clear_output()
            process_video_file()

    def on_benchmark_test(b):
        with output:
            clear_output()
            benchmark_system_performance()

    def on_advanced_demo(b):
        with output:
            clear_output()
            demonstrate_advanced_features()

    # Connect handlers
    test_detection_btn.on_click(on_detection_test)
    test_simulation_btn.on_click(on_simulation_test)
    test_video_btn.on_click(on_video_test)
    benchmark_btn.on_click(on_benchmark_test)
    advanced_demo_btn.on_click(on_advanced_demo)

    # Layout
    button_row1 = widgets.HBox([test_detection_btn, test_simulation_btn, test_video_btn])
    button_row2 = widgets.HBox([benchmark_btn, advanced_demo_btn])

    test_suite = widgets.VBox([
        widgets.HTML("<h2>🧪 Traffic Management System Test Suite</h2>"),
        button_row1,
        button_row2,
        output
    ])

    return test_suite

# =============================================================================
# MAIN EXECUTION FOR TESTING
# =============================================================================

print("🧪 Traffic Management System - Testing & Examples Loaded!")
print("\nAvailable test functions:")
print("- process_uploaded_image(): Test with your own traffic images")
print("- simulate_realtime_processing(): Real-time simulation with synthetic data")
print("- process_video_file(): Analyze uploaded traffic videos")
print("- benchmark_system_performance(): Performance testing")
print("- demonstrate_advanced_features(): Advanced capabilities demo")

# Create and display interactive test suite
interactive_suite = create_interactive_test_suite()
display(interactive_suite)

# Quick start example
print("\n🚀 Quick Start Example:")
print("Run this to test the system with a simple example:")
print()
print("# Initialize and test system")
print("system = TrafficManagementSystem()")
print("print('System ready for testing!')")
```
